- 이미지에 커널을 적용하여 이미지를 blur처리할 수 있다.

- 이미지를 흐리게 만들면 노이즈 및 손상을 줄일 수 있다.

- 특정한 이미지에서 커널을 적용해 컨볼루션 계산하여 필터링을 진행할 수 있다.

직접 커널 만들기

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image =cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
size =4
kernel = np.ones((size,size),np.float32)/(size**2)
print(kernel)
'''
 The larger the kernel size, the more pixels are
included in the average, and the more the image is blurred.
'''
```

```python
dst =cv2.filter2D(image,-1,kernel)
plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show()
```

Basic Blurring

```python
image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

dst = cv2.blur(image, (10, 10))
print(dst.shape)
print(image.shape)
#dst는 blur처리를 한 이미지이다.
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```

Gaussian Blur

```python
import cv2

image = cv2.imread('gray_image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# kernel_size: 홀수
dst = cv2.GaussianBlur(image, (3, 3), 0)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
